```python
def seg(document, horizontal_projection, target_path):
    '''
        Segment the document into lines using horizontal projection.
        
        Args:
            document: (nxm)
                the document image to be segmented
                horizontal_projection: (nx1)
                the horizontal projection value of the document 
            path: string
                the path of the segmented lines to be save
            
    '''
    #use cv2.reduce()
    #https://stackoverflow.com/questions/54285839/how-to-construct-horizontal-projection-of-binary-image-in-opencv
    
    line_start = []
    line_end = []

    is_start = True
    selected = False 
    is_append = False 
    first = 0
    second = 0

    min_gap = 5 
    avg_line_height = 0
    avg_gap = 0 #gap_number usually = line_number - 1 
#     total_lines = 0 
    current_line = -1 #line number starts from 0 
    
    #every time "end" update? 
    #line_height = end[i] - start[i]
    #gap = start[i+1] - end[i]
    
    
    for index, line in enumerate(horizontal_projection):
        if line == 0 and not selected:
            first = index
            selected = True
        elif line != 0 and selected :
            second = index - 1 
            mid = int((first + second) /2 )
            selected = False 

            if is_start:
                gap = -1 # declare
                try :
                    gap = index - line_end[current_line ]
                except IndexError as e: 
                    pass
                if gap < min_gap: 
                    is_start = False
                else: 
                    current_line += 1  
                    line_start.append(mid)
                    is_start = False
                    is_append = True 
            else:
                if not is_append: 
                    line_end.append(mid)
                    is_start = True

                    current_line += 1 
                
                else: 
                    line_end[-1] = index 
                    is_append = False 
                    is_start = False  
                    
       
    print(len(line_start), len(line_end))
    try:
        os.makedirs(target_path)
    except FileExistsError as e: 
        shutil.rmtree(target_path, ignore_errors = True)
        os.makedirs(target_path, exist_ok = True)
    except OSError as e:  
        if e.errno != os.errno.EEXIST:
            raise   
            
    print(line_start, line_end)
    for index, (start, end) in enumerate(zip(line_start, line_end)):
        print(start, end)
        cv2.imwrite(f'{target_path}\\Line{index}.png', document[start:end, :])
```
